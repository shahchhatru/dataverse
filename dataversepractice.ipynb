```python
#abstract_list[0].split()

def clean_text(text_list):
    corpus_l=[]
    for j in range(0,len(text_list)):
        var_clean=re.sub('[^a-zA-z]',' ',text_list[j])
        var_clean=var_clean.lower()
        var_clean=var_clean.split()
        var_clean=[lemmatizer.lemmatize(word) for word in var_clean if word not in set(stopwords.words('english'))]
        var_clean=" ".join(var_clean)
        corpus_l.append(var_clean)
        if j %10000==0:
            print(j,corpus_l[j])
    return corpus_l
```

## Building a Vocabulary


```python
SOS_token=0 # it signifies start of sentence it will be passed as first input to decoder

EOS_token=1 # it signifies end of sentence ,the decoder will stop predicting words once it outputs EOS

class Lang:
    def __init__(self,name):
        self.name=name   #name of our model generally in language translation task there are  two languages so we identify it with name =English for 
                         # english language and fr for french but here we here we have only one language so we assign name ="caption" or if you want 
                         #you can simply remove this parameter
        self.word2index={} # a dictionary that maps word to it's corresponding index

        self.word2count={} # a dictionary that stores the number of occurence of word in the word it can be used to eliminate rare words(words with few counts)

        self.index2word={
            0: "SOS", 1: "EOS"
        }
        self.n_words=2     # a integer variable that stores total no of word in language.
                           #initial value is 2 because there are already SOS and EOS token
    def addSentence(self,sentence):
        for word in sentence.split(' '):
            self.addWord(word)

    def addWord(self,word):
        if word not in self.word2index:
            self.word2index[word]=self.n_words # add word as key in dict with current value of n_words 
                                               #current value of n_word acts as a new index value 
            self.word2count[word]=1            #initial count of new word is always 1

            self.index2word[self.n_words]=word #add word and it's index to index2word dictionary

            self.n_words+=1                    # a new word is added to the language so update it.

        else:
            self.word2count[word]+=1           #since the word was already present we just increment it's no of occurence.

    def addSentenceList(self,sentence_list): #adds a list of sentence to the language
        for sentence in sentence_list:
            self.addSentence(sentence)

    def indexesFromSentence(self, sentence):
        return [self.word2index[word] for word in sentence.split(' ')]


    def VectorFromSentence(self, sentence):
        indexes = self.indexesFromSentence(sentence)
        indexes.append(EOS_token)
        return np.array(indexes, dtype=np.unit).view(-1, 1)
```
